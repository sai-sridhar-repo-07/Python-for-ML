```python
import sqlite3
```

```python
connection = sqlite3.connect('Ã©xample.db')
connection
```

```python
cursor = connection.cursor()
```

```python
cursor.execute('''
Create Table If Not Exists employees(
    id Integer Primary Key,
    name Text Not Null,
    age Integer,
    department Text 
    )
''')

connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')
```

```python
cursor.execute('''
Insert Into employees(name,age,department)
               values('Tarra',32,'Data Scientist')
''')


cursor.execute('''
Insert Into employees(name,age,department)
               values('Sai',23,'Engineer')
''')

cursor.execute('''
Insert Into employees(name,age,department)
               values('Sri',24,'CEO')
               ''')

connection.commit()
```

```python
cursor.execute('''
Select * from employees
''')

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
UPDATE employees 
    Set age = 12
    where name="Sai"
               
''')

connection.commit()
```
